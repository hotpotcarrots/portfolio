    // A list (array) of students
    let students = ["Aneesh", "Perry", "Moiz"];  // TODO: change these names

    // A list of scores
    let scores = [85, 90, 87];  // TODO: change these scores

    // String to build output
    let out = "";

    // 1. Print roster
    out += "<p><strong>Students:</strong> " + students.join(", ") + "</p>";

    // 2. Print first student and score
    out += "<p><strong>First student:</strong> " + students[0] + " - " + scores[0] + "</p>";

    // 3. Add a new student + score
    students.push("Tanay");
    scores.push(95);

    out += "<p><strong>After adding Tanay:</strong> " + students.join(", ") + "</p>";

    // 4. Remove a student who transfers out
    let index = students.indexOf("Moiz");
    if (index !== -1) {
      students.splice(index, 1);
      scores.splice(index, 1);
    }

    out += "<p><strong>After removing Moiz:</strong> " + students.join(", ") + "</p>";

    // 5. Curve all scores by +5
    for (let i = 0; i < scores.length; i++) {
      scores[i] += 5;
    }

    // 6. Print each student with score
    out += "<h3>Final Scores</h3>";
    for (let i = 0; i < students.length; i++) {
      out += "<p>" + students[i] + " - " + scores[i] + "</p>";
    }

    // Show results
    document.getElementById("output").innerHTML = out;
  </script>

<script>
  (function() {
    // isolate variables so they don't conflict with other cells
    let foods = ["pizza", "burger", "pasta"];
    let foodScores = [82, 93, 89];
    let out = "";

    out += "<p><strong>Foods:</strong> " + foods.join(", ") + "</p>";
    out += "<p><strong>First food:</strong> " + foods[0] + " - " + foodScores[0] + "</p>";

    foods.push("sushi");
    foodScores.push(95);
    out += "<p><strong>After adding sushi:</strong> " + foods.join(", ") + "</p>";

    let idx = foods.indexOf("pasta");
    if (idx !== -1) {
      foods.splice(idx, 1);
      foodScores.splice(idx, 1);
    }
    out += "<p><strong>After removing pasta:</strong> " + foods.join(", ") + "</p>";

    for (let i = 0; i < foodScores.length; i++) {
      foodScores[i] += 5;
    }

    out += "<h3>Final Scores</h3>";
    for (let i = 0; i < foods.length; i++) {
      out += "<p>" + foods[i] + " - " + foodScores[i] + "</p>";
    }

    document.getElementById("output-food").innerHTML = out;
  })();
</script>
